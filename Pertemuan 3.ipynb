**Fungsi aktivasi**
merupakan komponen penting dalam jaringan syaraf tiruan. jadi fungsi ini menentukan apakah neuron itu  harus diaktifkan/menghasilkan nilai atau tidak, berdasarkan bobot masukkannya.

## Sigmoid activation function ##
$$
  f(x) = \frac{1}{1+e^x}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def rumus_sigmoid(x):
  return 1/ (1+np.exp(-x))

x = np.linspace(-5,5,100)
y = rumus_sigmoid(x)

plt.plot(x,y)
plt.xlabel('input')
plt.ylabel('Output')
plt.title('Sigmoid Activation Function')
plt.grid(True) #biar ada kotak/gridnya
plt.show() #untuk menampilkan hasil

#semakin jauh range-nya berarti semakin tegak fungsi aktivasinya
```
